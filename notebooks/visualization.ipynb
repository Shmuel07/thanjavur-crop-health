# 03 — Visualization & Portfolio Output

Generates the final charts and summary table for the portfolio.

> **Note:** The hero map (NDVI raster over constituency boundaries) was
> produced in QGIS using real Sentinel-2 imagery and exported via Print Layout.
> This notebook generates the supporting ranking chart and summary statistics.

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

with open('../data/processed/ndvi_per_constituency.json') as f:
    data = json.load(f)
df = pd.DataFrame(data['constituencies']).sort_values('mean_ndvi', ascending=False)

print('=== THANJAVUR CROP HEALTH REPORT ===')
print(f"Date: {data['date']}  |  Source: {data['source']}")
print(f"Method: {data['method']}")
print()
print(f"{'Rank':<5}{'Constituency':<22}{'NDVI':>8}{'Max':>8}  Class")
print('-'*58)
for i,(_, r) in enumerate(df.iterrows(),1):
    print(f"{i:<5}{r.constituency:<22}{r.mean_ndvi:>8.4f}{r.max_ndvi:>8.3f}  {r.crop_health}")
print()
print(f"District mean NDVI: {df.mean_ndvi.mean():.4f}")
print(f"Seasonal context: {data['note']}")
```
