# 01 — Data Preparation

**Project:** Thanjavur Crop Health Monitoring using Sentinel-2 NDVI

## Data Sources

| Dataset | Source | Date | Use |
|---|---|---|---|
| Sentinel-2 L2A | ESA Copernicus Browser | 25 March 2026 | NDVI calculation |
| Constituency boundaries | Tamil Nadu Assembly Constituency (TNGIS) | 2024 | Zonal statistics |

## Sentinel-2 Bands Used
- **B04 (Red, 665nm)** — absorbed by chlorophyll
- **B08 (NIR, 842nm)** — reflected by healthy vegetation
- **Resolution:** 10m × 10m per pixel
- **Tile:** T44PKS (+ adjacent tiles for full district coverage)

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load NDVI zonal statistics from QGIS
with open('../data/processed/ndvi_per_constituency.json') as f:
    data = json.load(f)

print(f"Source: {data['source']}")
print(f"Date: {data['date']}")
print(f"Method: {data['method']}")
print(f"Note: {data['note']}")
print(f"\nConstituencies: {len(data['constituencies'])}")
```

```python
df = pd.DataFrame(data['constituencies'])
print(df[['constituency','mean_ndvi','max_ndvi','crop_health']].to_string(index=False))
```

## NDVI Value Reference

| NDVI Range | Meaning | Expected Land Cover |
|---|---|---|
| 0.50 – 1.00 | Dense healthy vegetation | Active rice / sugarcane |
| 0.30 – 0.50 | Moderate vegetation | Early stage crops / mixed |
| 0.10 – 0.30 | Sparse vegetation | Stressed / post-harvest |
| 0.00 – 0.10 | Bare / fallow | Resting fields |
| < 0.00 | Water / built-up | Tanks, rivers, urban |
